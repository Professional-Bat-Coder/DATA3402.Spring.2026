1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter.

```python
class Counter:
    def __init__(self, maximum):
        self.maximum = maximum
        self.value = 0

    def increment(self):
        if self.value >= self.maximum:
            print("Error: Counter already at maximum value!")
        else:
            self.value += 1

    def reset(self):
        self.value = 0

    def get_value(self):
        return self.value
```

```python
c = Counter(3)

print(c.get_value())  # 0

c.increment()
c.increment()
print(c.get_value())  # 2

c.increment()
print(c.get_value())  # 3

c.increment()  # should print error

c.reset()
print(c.get_value())  # 0
```

2. Copy and paste your solution to question 1 and modify it so that all the data held by the counter is private. Implement functions to check the value of the counter, check the maximum value, and check if the counter is at the maximum.

```python
class Counter:

    def __init__(self, maxval):
        self.__max = maxval
        self.__count = 0

    def increment(self):
        if self.__count == self.__max:
            print("cant go higher")
        else:
            self.__count = self.__count + 1

    def reset(self):
        self.__count = 0

    # check current value
    def value(self):
        return self.__count

    # check maximum
    def maximum(self):
        return self.__max

    # are we at max
    def at_max(self):
        if self.__count == self.__max:
            return True
        else:
            return False
```

```python
class Rectangle:

    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    # calculations
    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2*self.__length + 2*self.__width

    # accessors
    def get_length(self):
        return self.__length

    def get_width(self):
        return self.__width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y
```

4. Implement a class to represent a circle, holding the radius and $x$ and $y$ coordinates of center of the object. Implement functions that compute the area and perimeter of the rectangle. Make all data members private and privide accessors to retrieve values of data members.

```python
import math

class Circle:

    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    # calculations
    def area(self):
        return math.pi * self.__radius * self.__radius

    def perimeter(self):   # circumference
        return 2 * math.pi * self.__radius

    # accessors
    def get_radius(self):
        return self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object.

```python
class Shape:

    def area(self):
        raise NotImplementedError("area not implemented")

    def perimeter(self):
        raise NotImplementedError("perimeter not implemented")

    def get_x(self):
        raise NotImplementedError("get_x not implemented")

    def get_y(self):
        raise NotImplementedError("get_y not implemented")

    def points(self):
        raise NotImplementedError("points not implemented")
```

```python
class Rectangle(Shape):

    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2*self.__length + 2*self.__width

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def points(self):
        pts = []
        steps = 4  # 4 points per side -> 16 total

        for i in range(steps):
            t = i/(steps-1)
            pts.append((self.__x + t*self.__length, self.__y))  # bottom
            pts.append((self.__x + self.__length, self.__y + t*self.__width)) # right
            pts.append((self.__x + (1-t)*self.__length, self.__y + self.__width)) # top
            pts.append((self.__x, self.__y + (1-t)*self.__width)) # left

        return pts[:16]
```

```python
import math

class Circle(Shape):

    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y

    def area(self):
        return math.pi * self.__radius**2

    def perimeter(self):
        return 2*math.pi*self.__radius

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def points(self):
        pts = []
        n = 16
        for i in range(n):
            ang = 2*math.pi*i/n
            px = self.__x + self.__radius*math.cos(ang)
            py = self.__y + self.__radius*math.sin(ang)
            pts.append((px,py))
        return pts
```

```python
import math

class Triangle(Shape):

    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y

    def area(self):
        return 0.5*self.__base*self.__height

    def perimeter(self):
        hyp = math.sqrt(self.__base**2 + self.__height**2)
        return self.__base + self.__height + hyp

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def points(self):
        pts = []
        steps = 5

        A = (self.__x, self.__y)
        B = (self.__x+self.__base, self.__y)
        C = (self.__x, self.__y+self.__height)

        def interp(P,Q,t):
            return (P[0]+t*(Q[0]-P[0]), P[1]+t*(Q[1]-P[1]))

        for i in range(steps):
            t=i/(steps-1)
            pts.append(interp(A,B,t))
            pts.append(interp(B,C,t))
            pts.append(interp(C,A,t))

        return pts[:16]
```

```python
class Shape:

    def area(self):
        raise NotImplementedError

    def perimeter(self):
        raise NotImplementedError

    def points(self):
        raise NotImplementedError

    def contains(self, x, y):
        raise NotImplementedError
```

```python
def contains(self, x, y):
    if (x >= self.__x and x <= self.__x + self.__length and
        y >= self.__y and y <= self.__y + self.__width):
        return True
    else:
        return False
```

```python
def contains(self, x, y):
    dx = x - self.__x
    dy = y - self.__y
    dist = math.sqrt(dx*dx + dy*dy)

    if dist <= self.__radius:
        return True
    else:
        return False
```

```python
def contains(self, x, y):

    # shift coordinates
    x0 = x - self.__x
    y0 = y - self.__y

    if x0 < 0 or y0 < 0:
        return False
    if x0 > self.__base or y0 > self.__height:
        return False

    # line test
    if y0 <= (-self.__height/self.__base)*x0 + self.__height:
        return True
    else:
        return False
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
def overlaps(self, other):

    for (px, py) in self.points():
        if other.contains(px, py):
            return True

    for (px, py) in other.points():
        if self.contains(px, py):
            return True

    return False
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again.

```python
class RasterDrawing:

    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def paint(self):
        print("Painting Drawing:")
        for s in self.shapes:
            print("-", type(s).__name__, "at", s.get_x(), s.get_y())
            print("  area:", s.area(), "perimeter:", s.perimeter())
            print("  points:", s.points())
        print()
```

```python
class RasterDrawing:

    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def remove_shape(self, shape):
        if shape in self.shapes:
            self.shapes.remove(shape)

    def paint(self):
        print("Painting Drawing:")
        for s in self.shapes:
            print("-", type(s).__name__, "at", s.get_x(), s.get_y())
            print("  area:", s.area(), "perimeter:", s.perimeter())
            print("  points:", s.points())
        print()

    # --- new save/load methods ---
    def save(self, filename):
        with open(filename, "w") as f:
            for s in self.shapes:
                f.write(repr(s) + "\n")

    @staticmethod
    def load(filename):
        rd = RasterDrawing()
        with open(filename,"r") as f:
            for line in f:
                line = line.strip()
                if line:
                    obj = eval(line)  # recreate object
                    rd.add_shape(obj)
        return rd
```

```python
# create a drawing
r = Rectangle(4,3,0,0)
c = Circle(2,5,4)
t = Triangle(3,4,7,1)

drawing = RasterDrawing()
drawing.add_shape(r)
drawing.add_shape(c)
drawing.add_shape(t)

# paint
drawing.paint()

# save to file
drawing.save("drawing.txt")

# load from file
loaded_drawing = RasterDrawing.load("drawing.txt")

# paint loaded drawing
loaded_drawing.paint()
```
